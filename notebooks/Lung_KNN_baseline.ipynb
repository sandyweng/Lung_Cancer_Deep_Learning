```python
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array, array_to_img
%matplotlib inline
import glob
import numpy as np
```

```python
IMG_DIM = (150, 150)

train_files = glob.glob('Binary_Train/*')
train_imgs = [img_to_array(load_img(img, target_size=IMG_DIM)) for img in train_files]
train_imgs = np.array(train_imgs)
train_labels = [fn.split('\\')[1].split('.')[0].strip() for fn in train_files]


print('Train dataset shape:', train_imgs.shape)
```

```python
train_labels
```

```python
number_labels = []
for label in train_labels:
    if 'lungaca' in label:
        number_labels.append(0)
    else:
        number_labels.append(1)
```

```python
IMG_DIM = (150, 150)

test_files = glob.glob('Binary_Test/*')
test_imgs = [img_to_array(load_img(img, target_size=IMG_DIM)) for img in test_files]
test_imgs = np.array(test_imgs)
test_labels = [fn.split('\\')[1].split('.')[0].strip() for fn in test_files]
```

```python
test_labels
```

```python
num_test_labels = []
for label in test_labels:
    if 'lungaca' in label:
        num_test_labels.append(0)
    else:
        num_test_labels.append(1)
```

```python
num_test_labels
```

```python
test_imgs.shape
```

```python
train_imgs_reshaped = train_imgs.reshape(train_imgs.shape[0], 150*150*3)
train_imgs_reshaped.shape
```

```python
test_imgs_reshaped = test_imgs.reshape(test_imgs.shape[0], 150*150*3)
test_imgs_reshaped.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(train_imgs_reshaped, number_labels)
```

```python
y_pred = knn.predict(test_imgs_reshaped)
```

```python
print('accuracy: %f' % (np.mean(y_pred == num_test_labels)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(num_test_labels, y_pred))
print(classification_report(num_test_labels, y_pred))
```
