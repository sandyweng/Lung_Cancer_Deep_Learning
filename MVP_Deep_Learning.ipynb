# Lung Cancer Detection 

The goal of this project is to create a CNN model to predict whether a microscopic image of a lung tissue has lung cancer or not. 

Out of the 15,000 images in my dataset, I selected 750 images to run a small CNN model on my local machine. The train and validation accuracy were 0.5333. The graphs show that the model isn't overfitting but the accuracy is low because of the size of the data. 

After using feature extraction on VGG16 into fully connected layers, my training and validation accuracy increased to about 0.99 and 0.98. I will continue to fine-tune this model and try another pre-trained model to see if I can boost the accuracy score.
